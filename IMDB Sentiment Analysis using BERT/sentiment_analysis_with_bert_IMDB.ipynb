```python
df_train = pd.read_excel("IMDB-Movie-Reviews-Large-Dataset-50k/train.xlsx",dtype = str)
```

```python
df_test= pd.read_excel("IMDB-Movie-Reviews-Large-Dataset-50k/test.xlsx",dtype = str)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train.head()
```

```python
(X_train,y_train),(X_test,y_test),preprocess = text.texts_from_df(train_df = df_train,
                                                                  text_column = "Reviews",
                                                                  label_columns = "Sentiment",
                                                                  val_df = df_test,
                                                                  maxlen = 256,
                                                                  preprocess_mode = "bert")
```

```python
model = text.text_classifier(name = "bert",train_data = (X_train,y_train),preproc = preprocess)
```

```python
learner = ktrain.get_learner(model = model,train_data = (X_train,y_train),
                   val_data = (X_test,y_test),batch_size = 16)
```

```python
learner.fit_onecycle(lr = 2e-5,epochs = 3)
```

```python
predictor = ktrain.get_predictor(learner.model,preprocess)
```

```python
sentiments = ["this movie was awful,but acting was okay",
             "I WILL KILL YOU",
             "The film is realy sucked,there is not plot and acting was bad ",
             "was a beautiful movie,great plot .acting was very good",
             "what a nice movie , I like it"]
```

```python
predictor.predict(sentiments)
```

```python
predictor.predict(sentiments,return_proba = True)
```
